```python
data = pd.read_csv('documents/ТМО Лаб.1/wine.txt', sep=",")
```

```python
# Проверим наличие пустых значений
# Цикл по колонкам датасета
for col in data.columns:
    # Количество пустых значений - все значения заполнены
    temp_null_count = data[data[col].isnull()].shape[0]
    print('{} - {}'.format(col, temp_null_count))
```

# 3) Визуальное исследование датасета

```python
# Гистограммы для всех признаков
data.hist(bins=30, figsize = (15,7))
```

```python
# Диаграммы рассеяние для всех признаков
plt.figure(figsize=(12,6))
sns.pairplot(data)
```

```python
# Увеличенные диаграммы рассеяния для признаков, которые имеют зависимость с уровнем счастья
sns.jointplot(x = "Alcohol", y = "Malic Acid", kind="scatter", data = data)
```
