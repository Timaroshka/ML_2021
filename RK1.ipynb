## Характеристики датасета

```python
# импорт библиотек
from sklearn.datasets import load_boston
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set(style="ticks")
```

```python
# преобразование и загрузка данных 
boston = load_boston()
data = pd.DataFrame(boston.data, columns=boston.feature_names)
```

## Корреляция

```python
corr_matrix = data.corr()
```

```python
data.corr()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(corr_matrix, annot=True, fmt='.3f')
```

```python
# Корреляционная матрица содержит коэффициенты корреляции между всеми парами признаков.
Корреляционная матрица симметрична относительно главной диагонали. На главной диагонали расположены единицы (корреляция признака самого с собой). На основе корреляционной матрицы можно сделать следующие выводы:
1) B и DIS негативно влияют на корреляционную матрицу, что мешает точной оценке данных. Их стоит удалить.
2) TAX наиболее сильно коррелирует с RAD.
3) CHAS коррелирует с INDUS, NOX, RM, AGE и B.
```
